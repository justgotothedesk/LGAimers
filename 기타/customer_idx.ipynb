```python
train_file_path = '/content/drive/MyDrive/LGdata/train.csv'

df = pd.read_csv(train_file_path)
```

```python
df_customer_idx = df.dropna(subset=['customer_idx'])

converted_counts_customer_idx = df_customer_idx.groupby('customer_idx')['is_converted'].sum().reset_index()
total_counts_customer_idx = df_customer_idx.groupby('customer_idx')['is_converted'].count().reset_index()

ratio_df_customer_idx = pd.merge(converted_counts_customer_idx, total_counts_customer_idx, on='customer_idx', how='left')
ratio_df_customer_idx['conversion_ratio'] = ratio_df_customer_idx['is_converted_x'] / ratio_df_customer_idx['is_converted_y']
count = 0

for index, row in ratio_df_customer_idx.iterrows():
    total = int(row['is_converted_y'])
    converted = int(row['is_converted_x'])
    ratio = row['conversion_ratio'] * 100
    if total > 3:
      count += 1
      print(f"{row['customer_idx']} 별 is_converted 비율: {ratio:.2f}%, 전체 갯수: {total}, is_converted된 갯수: {converted}")

print("\n\n동일한 customer_idx가 3보다 큰 갯수 : ", count)
print("그 비율 : ", count/52599*100)
```
